```python
import pandas as pd
```

```python
for data_type in ['train', 'dev', 'test']:
    df = pd.read_csv(f'./metadata/{data_type}.csv')
    print('Data type:', data_type)
    print('# utterances (k) / duration (hr): {:.1f}k / {:.1f}h'.format(len(df)/1000, df['duration'].sum() / 3600))
    # get the number of samples for each accent and total duration of audio for each accent
    stats = df.groupby('accent').agg({'audio_path': 'count', 'duration': 'sum'}).reset_index()
    stats.columns = ['accent', 'num_samples', 'total_duration_sec']
    stats['total_duration_hr'] = stats['total_duration_sec'] / 3600
    # print(stats)
    print("=="*50)
```
